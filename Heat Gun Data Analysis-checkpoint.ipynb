```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import StaleElementReferenceException, ElementClickInterceptedException
import time
import pandas as pd
from selenium.webdriver import ActionChains

website = 'https://www.amazon.com/s?k=heat+gun&crid=1P0EPCC5JJ13D&sprefix=%2Caps%2C228&ref=nb_sb_ss_recent_1_0_recent'
path = 'C:\\Users\\Wishes Lawrence\\Documents\\Chromedriver\\chromedriver-win64\\chromedriver.exe'
chrome_options = Options()
chrome_options.add_argument("--disable-notifications")
driver = webdriver.Chrome(options=chrome_options)
driver.maximize_window()
driver.get(website)
driver.implicitly_wait(5)

# Open the website and wait for it to load
time.sleep(10)

try:
    actions = ActionChains(driver)
    
    # Scroll into view before clicking the region/country element
    country_element = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.ID, 'icp-nav-flyout'))
    )
    actions.move_to_element(country_element).perform()

    # Click the region/country change link
    change_country = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, '//span[contains(@class,"icp-nav-flag")]'))
    )
    change_country.click()

    # Select the desired country (e.g., India)
    india_option = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, '//a[@role="button" and contains(text(), "India - INR")]'))
    )
    india_option.click()
    
    # Click on "Go to website"
    goto_website = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.ID, 'icp-save-button-announce'))
    )
    goto_website.click()

except Exception as e:
    print(f"Error changing country: {e}")

# Search for the keyword
search_box = driver.find_element(By.ID, 'twotabsearchtextbox')
search_box.send_keys("heat gun")
submit_keyword = driver.find_element(By.ID, 'nav-search-submit-button')
submit_keyword.click()

# Wait for the webpage to load
driver.implicitly_wait(5)

# Lists to store extracted data
product_name = []
product_price = []
product_ratings = []
product_ratings_num = []
product_link = []
product_bought = []

# Start pagination process
current_page = 1
last_age=7
while current_page<=last_page:
    print(f"Scraping page {current_page}")

    # Wait for products to load
    items = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, '//div[@data-component-type="s-search-result"]'))
    )
    
    for item in items:
        try:
            name = item.find_element(By.XPATH, './/h2/a/span').text
            product_name.append(name)
        except Exception as e:
            product_name.append("N/A")
        
        try:
            price = item.find_element(By.XPATH, './/span[@class="a-price-whole"]').text
            product_price.append(price)
        except Exception as e:
            product_price.append("N/A")

        try:
            ratings = item.find_element(By.XPATH, './/span[@class="a-icon-alt"]').text
            product_ratings.append(ratings)
        except Exception as e:
            product_ratings.append("N/A")

        try:
            ratings_num = item.find_element(By.XPATH, './/span[@class="a-size-base s-underline-text"]').text
            product_ratings_num.append(ratings_num)
        except Exception as e:
            product_ratings_num.append("N/A")

        try:
            how_many_bought = item.find_element(By.XPATH, './/span[@class="a-size-base a-color-secondary"]').text
            product_bought.append(how_many_bought)
        except Exception as e:
            product_bought.append("N/A")

        try:
            link = item.find_element(By.XPATH, './/h2/a').get_attribute("href")
            product_link.append(link)
        except Exception as e:
            product_link.append("N/A")

    # Check if there is a next page
    current_page=current_page+1
    try:
        next_button=driver.find_element(By.XPATH,'//a[contains(@class,"s-pagination-next ")]')
        next_button.click() 
    except:
        pass
        current_page += 1
        

# Creating DataFrame
df = pd.DataFrame({
    'product_name': product_name,
    'product_price': product_price,
    'product_ratings': product_ratings,
    'product_ratings_num': product_ratings_num,
    'product_link': product_link,
    'product_bought': product_bought
})

print(df)
df.to_excel('amazon.xlsx', index=False)
# Close the WebDriver
driver.quit()
```
