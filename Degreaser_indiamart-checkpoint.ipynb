```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import StaleElementReferenceException, ElementClickInterceptedException
from selenium.webdriver.common.action_chains import ActionChains

# Initialize WebDriver
website = 'https://dir.indiamart.com/search.mp?ss=degreaser&mcatid=12735&catid=519&prdsrc=1&stype=attr=1|attrS&res=RC4'
path = 'C:\\Users\\Wishes Lawrence\\Documents\\Chromedriver\\chromedriver-win64\\chromedriver.exe'

chrome_options = Options()
chrome_options.add_argument("--disable-notifications")
driver = webdriver.Chrome(options=chrome_options)
driver.maximize_window()
driver.get(website)

# Wait for the elements to be visible
wait = WebDriverWait(driver, 20)
names = []
prices = []
units = []
company = []
comp_address = []

degreaser_links = wait.until(EC.visibility_of_all_elements_located((By.XPATH, '//span[@class="elps elps2 p10b0 fs14 tac mListNme"]/a')))

for degreaser in degreaser_links:
    try:
        # Scroll to the element to ensure it's in the viewport
        ActionChains(driver).move_to_element(degreaser).perform()
        
        # Try to click the element
        degreaser.click()
        print(degreaser.text)
    except ElementClickInterceptedException:
        # Handle the pop-up or overlay that intercepts the click
        try:
            # Add code to close the pop-up or overlay here
            # For example, if there's a close button, click it:
            close_button = driver.find_element_by_xpath("//button[@id='popup-close-button']")
            close_button.click()
        except:
            # If there's no close button or the pop-up handling fails, you may need to wait for it to disappear
            WebDriverWait(driver, 10).until(EC.invisibility_of_element_located((By.XPATH, "//div[@id='popup-overlay']")))
            
        # Try clicking the element again
        degreaser.click()
        print(degreaser.text)
    
    try:
        name = wait.until(EC.visibility_of_element_located((By.XPATH, '//h1[@class="bo center-heading"]')))
        names.append(name.text)
        print(name.text)
    except:
        names.append("NA")
    try:
        price = wait.until(EC.visibility_of_element_located((By.XPATH, '//span[@class="prc-tip bo price-unit"]')))
        prices.append(price.text)
        print(price.text)
    except:
        prices.append("NA")

    try:
        unit = wait.until(EC.visibility_of_element_located((By.XPATH, '//span[@class="fs15 pcl76"]')))
        units.append(unit.text)
        print(unit.text)
    except:
        units.append("NA")

    try:
        company_name = wait.until(EC.visibility_of_element_located((By.XPATH, '//h2[@class="fs15 bo"]')))
        company.append(company_name.text)
        print(company_name.text)
    except:
         company.append("NA")
    try:
        company_address = wait.until(EC.visibility_of_element_located((By.XPATH, '//div[@class="fs12 color1 dsf addrs plhn"]')))
        comp_address.append(company_address.text)
        print(company_address.text)
    except:
         comp_address.append("NA")

# Close the WebDriver
driver.quit()

# Now, you can work with the extracted data stored in the lists: names, prices, units, company, comp_address
```
